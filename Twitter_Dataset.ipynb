```python
useless_words = nltk.corpus.stopwords.words("english") + list (string.punctuation)
```

```python
useless_words
```

```python
def build_bag_of_words_features_filtered(words):
    return {
        word: 1 for word in words \
        if not word in useless_words}
```

```python
build_bag_of_words_features_filtered(twitter_words)
```

```python
twitter_words = twitter_samples.strings()
len (twitter_words)/1e6
```

```python
filtered_words = [word for word in twitter_samples.strings() if not  word in useless_words]
type (filtered_words)
```

```python
len (filtered_words)/1e6
```

```python
from collections import Counter
words_counter = Counter (filtered_words)
```

```python
most_common_words = words_counter.most_common() [:10]
```

```python
most_common_words
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
sorted_word_counts = sorted (list(words_counter.values()), reverse = True)
plt.loglog(sorted_word_counts)
plt.ylabel("Freq")
plt.xlabel("Word Rank");
```
